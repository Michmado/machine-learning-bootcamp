**1. Import bibliotek**


```python
import numpy as npp
import pandas as pd
import sklearn
```

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}

df_raw = pd.DataFrame(data=data)
df_raw
```

```python
df = df_raw.copy()
df.info()
```

**Zmiana typu danych i wstÄ™pna eksploracja**

```python
for col in ['size', 'color', 'gender', 'bought']:
  df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

```python
df.describe()
```

```python
df.describe().T
```

```python
df.describe(include=['category']).T
```

**LabelEncoder**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()  #tworzymi klasy instancje
le.fit(df['bought'])  # metoda fit dopasuje encoder do naszej zmiennej
le.transform(df['bought'])  # metoda transform mapuje i przetransferowywuje naszÄ… zmiennÄ…  juz do modelu
```

```python
le.fit_transform(df['bought']) #to samo co wyzej tylo w jednym razem
```

```python
le.classes_ # pozwala wyÅ›wietliÄ‡ tablicÄ™ z jakÄ… dokonuje siÄ™ mapowanie
```

```python
df['bought'] = le.fit_transform(df['bought'])
df
```

```python
le.inverse_transform(df['bought']) # pozwala no odwrÃ³cenie mapowania - wrÃ³cenie do satanu domyÅ›lnego
```
